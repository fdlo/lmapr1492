```python
#Telecharger le ficher CIF du materiau mp-1115 (PtSe2)
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter

with MPRester("NLKe1dAg7yznVObH") as m:
    # Create a structure for material id mp-1115 (PtSe2)
    structure = m.get_structure_by_material_id("mp-1115", True, True)

w = CifWriter(structure)
w.write_file("PtSe2_mp-1115.cif")
```

# Le difractogramme pour  *$2\theta \in [0°,90°]$*

```python
from pymatgen.analysis.diffraction import xrd
from pymatgen.analysis.diffraction import core
from matplotlib import pyplot as plt

#La radiation CuKa1 a une longueur d'onde de 1.5406
xrdc = xrd.XRDCalculator(wavelength='CuKa1')

#Obtention du pattern de diffraction de notre structure:
diffpattern = xrdc.get_pattern(structure)

#On obtient un graphe qui nous montre l'intensité des pics en fonction de l'angle 2\theta
xrdc.get_plot(structure, annotate_peaks = False)
```

# Les trois premiers pics

```python
#Premier pic:
print("Angle du premier pic: " + str(diffpattern.x[0]) + "°")
print("Intensité du premier pic: " + str(diffpattern.y[0]))
print("Indices hkl et multiplicité du premier pic: " + str(diffpattern.hkls[0]) + "\n")

#Deuxieme pic:
print("Angle du deuxième pic: " + str(diffpattern.x[1]) + "°")
print("Intensité du deuxième pic: " + str(diffpattern.y[1]))
print("Indices hkl et multiplicité du deuxième pic: " + str(diffpattern.hkls[1]) + "\n")

#Troisieme pic:
print("Angle du troisième pic: " + str(diffpattern.x[2]) + "°")
print("Intensité du troisième pic: " + str(diffpattern.y[2]))
print("Indices hkl et multiplicité du troisième pic: " + str(diffpattern.hkls[2]))
```
