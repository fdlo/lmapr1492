```python
#Telecharger le ficher CIF du materiau mp-1115 (PtSe2)
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter

with MPRester("NLKe1dAg7yznVObH") as m:
    # Create a structure for material id mp-1115 (PtSe2)
    structure = m.get_structure_by_material_id("mp-1115", True, True)

w = CifWriter(structure)
w.write_file("PtSe2_mp-1115.cif")
```

```python
from pymatgen.symmetry import analyzer
from numpy import dot
from IPython.display import Image

#SpacegroupAnalyser SGA:
SGA = analyzer.SpacegroupAnalyzer(structure)

SGO = SGA.get_space_group_operations()
print(SGO)

symop = SGA.get_symmetry_operations()
site = structure.sites
print("\nL'espèce, les coordonnées carthésiennes et les coordonnées fractionnaires des 3 atomes que nous allons analyser:")
print(site) 
```


# Opération de symétrie 1: plan mirroir (-y,-x,z)

```python
symop1 = symop[7] 
rot1 = symop1.rotation_matrix
transl1 = symop1.translation_vector
print("Matrice de rotation:")
print(rot1)
print("Vecteur de translation:")
print(transl1)
print("Elément et coordonées fractionnaires de départ de l'atome considéré:")
print(site[0].species, site[0].frac_coords)
print("Coordonnées fractionnaires après l'opération de symétrie:")
#produit matriciel
print(dot(rot1, site[0].frac_coords) + transl1)
```
